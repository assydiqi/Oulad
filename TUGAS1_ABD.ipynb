#Analisis pendaftaran dan penggunaan vle berdasarkanregion

Analisis ini bertujuan untuk mengetahui kapan siswa mendaftar dan mulai menggunakan VLE berdasarakan wilayah geografis

```python
# Baca dataset "studentRegistration.csv"
registration_data = pd.read_csv('/content/drive/MyDrive/TUGAS 1/studentRegistration.csv')
```

```python
# Baca dataset "studentVle.csv"
vle_data = pd.read_csv('/content/drive/MyDrive/TUGAS 1/studentVle.csv')
```

```python
# Baca dataset "studentInfo.csv" untuk mendapatkan kolom "region"
info_data = pd.read_csv('/content/drive/MyDrive/TUGAS 1/studentInfo.csv')[['id_student', 'region']]
```

```python
# Gabungkan dataset "studentRegistration.csv" dan "studentVle.csv" berdasarkan "id_student"
merged_data = pd.merge(registration_data, vle_data, on='id_student')
```

```python
# Gabungkan dengan dataset "studentInfo.csv" untuk mendapatkan informasi wilayah
merged_data = pd.merge(merged_data, info_data, on='id_student')
```

```python
merged_data.head()
```

```python
# Menampilkan informasi umum mengenai dataset
print(merged_data.info())
```

```python
# Statistik deskriptif
print(merged_data.describe())
```

```python
# Menampilkan jumlah nilai null dalam setiap kolom
print(merged_data.isnull().sum())
```

```python
# Menampilkan jumlah duplikat data
print("Jumlah Duplikat: ", merged_data.duplicated().sum())
```

```python
merged_datamerged_data = merged_data.drop_duplicates()
print("Jumlah Duplikat: ", merged_data.duplicated().sum())

print("Jumlah baris data setelah menghapus duplikat:", len(merged_data))
```

Analisis Data dan Membuat Grafik: Setelah dataset digabungkan, lakukan analisis data untuk memahami tren pendaftaran dan penggunaan VLE berdasarkan wilayah. Ini mungkin melibatkan penggunaan groupby dan visualisasi data menggunakan library seperti Matplotlib atau Seaborn.

```python
# Visualisasi distribusi data
plt.figure(figsize=(12, 6))
sns.histplot(data=merged_data, x='date_registration', kde=True)
plt.title('Distribusi Tanggal Pendaftaran')
plt.xlabel('Tanggal Pendaftaran')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=merged_data, x='date', kde=True)
plt.title('Distribusi Tanggal Penggunaan VLE')
plt.xlabel('Tanggal Penggunaan VLE')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Kelompokkan data berdasarkan wilayah (region) dan code_module, lalu hitung jumlah pendaftaran
grouped_data = merged_data.groupby(['region', 'code_module_x'])['id_student'].count().reset_index()

# Buat grafik tren pendaftaran berdasarkan wilayah dan code_module
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_data, x='region', y='id_student', hue='code_module_x')
plt.title('Tren Pendaftaran Berdasarkan Wilayah dan Code Module')
plt.xlabel('Region')
plt.ylabel('Jumlah Pendaftaran')
plt.xticks(rotation=45)
plt.legend(title='Code Module')
plt.show()
print("Jumlah pendaftaran pada setiap wilayah")
```

```python
# Kelompokkan data berdasarkan wilayah (region) dan code_module, lalu hitung jumlah pendaftaran
grouped_data = merged_data.groupby(['region', 'code_module_y'])['id_student'].count().reset_index()

# Buat grafik tren pendaftaran berdasarkan wilayah dan code_module
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_data, x='region', y='id_student', hue='code_module_y')
plt.title('Tren Pendaftaran Berdasarkan Wilayah dan Code Module')
plt.xlabel('Region')
plt.ylabel('Jumlah Pendaftaran')
plt.xticks(rotation=45)
plt.legend(title='Code Module')
plt.show()
```

```python
# Kelompokkan data berdasarkan wilayah (region), lalu hitung jumlah pendaftaran dan penggunaan VLE
grouped_data = merged_data.groupby('region').agg({'id_student': 'count', 'sum_click': 'sum'}).reset_index()

# Buat grafik scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=grouped_data, x='id_student', y='sum_click', hue='region', s=100)
plt.title('Korelasi Jumlah Pendaftaran dan Penggunaan VLE Berdasarkan Wilayah')
plt.xlabel('Jumlah Pendaftaran')
plt.ylabel('Jumlah Penggunaan VLE')
plt.legend(title='Region', loc='upper right')
plt.show()
```

Grafik Batang untuk Jumlah Pendaftaran per Wilayah

```python
# Kelompokkan data berdasarkan wilayah (region) dan hitung jumlah pendaftaran
pendaftaran_per_wilayah = merged_data.groupby('region')['id_student'].count().reset_index()

# Membuat grafik batang
plt.figure(figsize=(12, 6))
sns.barplot(data=pendaftaran_per_wilayah, x='region', y='id_student', palette='viridis')
plt.title('Jumlah Pendaftaran Siswa per Wilayah')
plt.xlabel('Wilayah')
plt.ylabel('Jumlah Pendaftaran')
plt.xticks(rotation=45)
plt.show()
```

Grafik Batang untuk Jumlah Penggunaan VLE per Wilayah

```python
# !
# Kelompokkan data berdasarkan wilayah (region) dan hitung jumlah penggunaan VLE
penggunaan_VLE_per_wilayah = merged_data.groupby('region')['sum_click'].sum().reset_index()

# Membuat grafik batang
plt.figure(figsize=(12, 6))
sns.barplot(data=penggunaan_VLE_per_wilayah, x='region', y='sum_click', palette='magma')
plt.title('Jumlah Penggunaan VLE per Wilayah')
plt.xlabel('Wilayah')
plt.ylabel('Jumlah Penggunaan VLE')
plt.xticks(rotation=45)
plt.show()
```

```python
# !
# Hitung statistik deskriptif jumlah pendaftaran per wilayah
statistik_pendaftaran = pendaftaran_per_wilayah['id_student'].describe()

# Hitung statistik deskriptif jumlah penggunaan VLE per wilayah
statistik_penggunaan_VLE = penggunaan_VLE_per_wilayah['sum_click'].describe()

# Tampilkan statistik deskriptif
print("Statistik Jumlah Pendaftaran per Wilayah:")
print(statistik_pendaftaran)
print("\nStatistik Jumlah Penggunaan VLE per Wilayah:")
print(statistik_penggunaan_VLE)

# Buat grafik untuk membandingkan pendaftaran dan penggunaan VLE per wilayah
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(data=pendaftaran_per_wilayah, x='region', y='id_student', palette='viridis')
plt.title('Jumlah Pendaftaran Siswa per Wilayah')
plt.xlabel('Wilayah')
plt.ylabel('Jumlah Pendaftaran')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(data=penggunaan_VLE_per_wilayah, x='region', y='sum_click', palette='magma')
plt.title('Jumlah Penggunaan VLE per Wilayah')
plt.xlabel('Wilayah')
plt.ylabel('Jumlah Penggunaan VLE')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Grafik ini memungkinkan Anda untuk membandingkan dua aspek yang berbeda, yaitu pendaftaran dan penggunaan VLE, dalam satu tampilan visual. Anda dapat memodifikasi kode ini sesuai dengan preferensi Anda untuk analisis lebih lanjut, seperti menghitung korelasi antara pendaftaran dan penggunaan VLE, atau membuat visualisasi tambahan.
